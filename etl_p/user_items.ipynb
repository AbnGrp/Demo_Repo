The following notebook contains an ETL process to which a file containing data from the video games platform STEAM was subjected.

```python
#Needed_Libraries

import pandas as pd
import numpy as np
import json
import re
from etl_functions import general_information, load_json_lines, duplicated_values, expand_json_column
```

(For the sake of practicality, we will take just a small section of the dataframe to keep going with our further analysis. We will have to keep this in mind when deriving any conclusions about the data).

```python
df_items1=df_items.loc[:900,:]
```

```python
df_items1.head(5)
```

```python
df_items1=expand_json_column(df_items1,"items")
```

```python
df_items1.head(5)
```

```python
df_items1=pd.melt(df_items1,id_vars=["user_id"],value_vars=list(range(1280)),value_name="items")
```

```python
df_items1.head(5)
```

```python
df_items1.dropna(subset=["items"],inplace=True)
```

```python
df_items1.head(5)
```

```python
general_information(df_items1)
```

```python
df_items1.dropna(subset=["user_id"],inplace=True)
```

```python
df_items1.head(5)
```

```python
user_id_1=df_items1["user_id"].copy()
```

```python
df_items1=df_items1["items"].apply(pd.Series,dtype="object")
```

```python
df_items1.head(5)
```

```python
general_information(df_items1)
```

```python
df_items1["user_id"]=user_id_1
```

```python
df_items1.drop(columns=["item_name"],inplace=True)
```
